**Experiment 5:**

WAP to train and evaluate a convolutional neural network using Keras Library to classify MNIST fashion dataset.

1. Input: 28×28 grayscale images with 1 channel → input shape: (28, 28, 1).

2. Convolutional Layers:

    * Block 1:
        * Conv2D with 64 filters (3×3), ReLU activation    
        * Batch Normalization
        * Conv2D with 64 filters (3×3), ReLU activation     
        * Batch Normalization
        * MaxPooling2D
        * Dropout: 30%

    * Block 2:
       * Conv2D with 128 filters, ReLU + BatchNorm
       * Conv2D with 128 filters, ReLU + BatchNorm
       * MaxPooling2D
       * Dropout: 40%

    * Block 3:
      * Conv2D with 256 filters, ReLU + BatchNorm
      * Conv2D with 256 filters, ReLU + BatchNorm
      * MaxPooling2D
      * Dropout: 40%

   * Block 4:
      * Conv2D with 512 filters, ReLU + BatchNorm
      * Conv2D with 256 filters, ReLU + BatchNorm
      * MaxPooling2D
      * Dropout: 40%

3. Fully Connected (Dense) Layers:

  * Dense Layer: 1024 neurons, ReLU + Dropout (50%)

  * Dense Layer: 512 neurons, ReLU + Dropout (50%)

  * Dense Layer: 256 neurons, ReLU + Dropout (50%)

  * Output Layer: 10 neurons (for 10 classes), softmax activation

4. Loss Function: Sparse Categorical Cross-Entropy

5. Optimizer: AdamW

  * Weight decay via L2 regularization (factor = 0.0001)

6. Learning Rate: 0.0005 (fixed)

  * Additional scheduling via ReduceLROnPlateau

7. Batch Sizes Used:

  * 128

  * 200

8. Epochs:

  * 150 epochs (early stopping may stop training earlier)

9. Metrics:

  * Train Accuracy

  * Train Loss

  * Test Accuracy

  * Test Loss

  * Confusion Matrix (on test set)

  * Classification Report: Precision, Recall, F1-score for each class

  * Plots: Train/Validation Accuracy and Loss over epochs



```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, regularizers, optimizers, callbacks
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

def load_data():
    (x_train, y_train), (x_test, y_test) = datasets.fashion_mnist.load_data()
    x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalize
    x_train = np.expand_dims(x_train, axis=-1)  # Add channel dimension
    x_test = np.expand_dims(x_test, axis=-1)
    return (x_train, y_train), (x_test, y_test)

def create_model(filter_size=3, reg=None, optimizer='adam'):
    model = models.Sequential([
        layers.Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=reg, input_shape=(28, 28, 1)),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2), padding='same'),
        layers.Dropout(0.3),

        layers.Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2), padding='same'),
        layers.Dropout(0.4),

        layers.Conv2D(256, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.Conv2D(256, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2), padding='same'),
        layers.Dropout(0.4),


        layers.Conv2D(512, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.Conv2D(256, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2), padding='same'),
        layers.Dropout(0.4),


        layers.Flatten(),
        layers.Dense(1024, activation='relu', kernel_regularizer=reg),
        layers.Dropout(0.5),
        layers.Dense(512, activation='relu', kernel_regularizer=reg),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu', kernel_regularizer=reg),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

def plot_loss_accuracy(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Loss Curve')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title('Accuracy Curve')

    plt.show()

def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=range(10), yticklabels=range(10))
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.show()

def train_and_evaluate(batch_size, filter_size, reg, optimizer, epochs):
    (x_train, y_train), (x_test, y_test) = load_data()
    model = create_model(filter_size, reg, optimizer)

    lr_scheduler = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
    early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)

    datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        rotation_range=10, width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True
    )
    datagen.fit(x_train)

    history = model.fit(datagen.flow(x_train, y_train, batch_size=batch_size),
                        epochs=epochs, validation_data=(x_test, y_test),
                        callbacks=[lr_scheduler, early_stopping])

    test_loss, test_acc = model.evaluate(x_test, y_test)
    train_loss = history.history['loss'][-1]
    train_acc = history.history['accuracy'][-1]

    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)

    optimizer_name = optimizer.__class__.__name__
    print(f"Batch Size: {batch_size}, Filter Size: {filter_size}, Optimizer: {optimizer_name}, Regularization: {reg}, Train Accuracy: {train_acc:.4f}, Train Loss: {train_loss:.4f}, Test Accuracy: {test_acc:.4f}, Test Loss: {test_loss:.4f}")

    plot_loss_accuracy(history)
    plot_confusion_matrix(y_test, y_pred_classes)
    print(classification_report(y_test, y_pred_classes))

    return batch_size, filter_size, optimizer_name, reg, train_acc, train_loss, test_acc, test_loss

# Define different hyperparameter combinations
hyperparams = [
    (200, 3, regularizers.l2(0.0001), optimizers.AdamW(learning_rate=0.0005), 150),
    (128, 3, regularizers.l2(0.0001), optimizers.AdamW(learning_rate=0.0005), 150)
]

# Store results
results = []
for params in hyperparams:
    results.append(train_and_evaluate(*params))

# Save results to CSV
df = pd.DataFrame(results, columns=['Batch Size', 'Filter Size', 'Optimizer', 'Regularization', 'Train Accuracy', 'Train Loss', 'Test Accuracy', 'Test Loss'])
df.to_csv('cnn_fashion_mnist_results.csv', index=False)
print("Results saved to cnn_fashion_mnist_results.csv")
```

  * Accuracy is also tracked during training and validation as a performance metric.

  * Optimizer (e.g., AdamW) is configured with a specified learning rate.

6. Training Loop
  * The model is trained on different sets of hyperparameters, as specified in a list:

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, regularizers, optimizers, callbacks
from sklearn.utils import class_weight
from tensorflow.keras.utils import to_categorical
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

def load_data():
    (x_train, y_train), (x_test, y_test) = datasets.fashion_mnist.load_data()
    x_train, x_test = x_train / 255.0, x_test / 255.0
    x_train = np.expand_dims(x_train, axis=-1)
    x_test = np.expand_dims(x_test, axis=-1)
    y_train = to_categorical(y_train, 10)
    y_test = to_categorical(y_test, 10)
    return (x_train, y_train), (x_test, y_test)

def create_model(filter_size=3, reg=None, optimizer='adam'):
    model = models.Sequential([
        layers.Conv2D(64, (filter_size, filter_size), activation='relu', padding='same', kernel_regularizer=reg, input_shape=(28, 28, 1)),
        layers.BatchNormalization(),
        layers.Conv2D(64, (filter_size, filter_size), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2), padding='same'),
        layers.Dropout(0.3),

        layers.Conv2D(128, (filter_size, filter_size), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.Conv2D(128, (filter_size, filter_size), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2), padding='same'),
        layers.Dropout(0.4),

        layers.Conv2D(256, (filter_size, filter_size), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.Conv2D(256, (filter_size, filter_size), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2), padding='same'),
        layers.Dropout(0.4),

        layers.Conv2D(512, (filter_size, filter_size), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.Conv2D(256, (filter_size, filter_size), activation='relu', padding='same', kernel_regularizer=reg),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2), padding='same'),
        layers.Dropout(0.4),

        layers.Flatten(),
        layers.Dense(1024, activation='relu', kernel_regularizer=reg),
        layers.Dropout(0.5),
        layers.Dense(512, activation='relu', kernel_regularizer=reg),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu', kernel_regularizer=reg),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])

    model.compile(optimizer=optimizer,
                  loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
                  metrics=['accuracy'])
    return model

def plot_loss_accuracy(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Loss Curve')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title('Accuracy Curve')

    plt.show()

def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=range(10), yticklabels=range(10))
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.show()

def train_and_evaluate(batch_size, filter_size, reg, optimizer, epochs):
    (x_train, y_train), (x_test, y_test) = load_data()
    model = create_model(filter_size, reg, optimizer)

    # Compute class weights using original labels
    y_train_labels = np.argmax(y_train, axis=1)
    class_weights_array = class_weight.compute_class_weight(class_weight='balanced',
                                                            classes=np.unique(y_train_labels),
                                                            y=y_train_labels)
    class_weights_dict = dict(enumerate(class_weights_array))

    lr_scheduler = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
    early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)

    datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        rotation_range=10, width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True
    )
    datagen.fit(x_train)

    history = model.fit(datagen.flow(x_train, y_train, batch_size=batch_size),
                        epochs=epochs,
                        validation_data=(x_test, y_test),
                        class_weight=class_weights_dict,
                        callbacks=[lr_scheduler, early_stopping])

    test_loss, test_acc = model.evaluate(x_test, y_test)
    train_loss = history.history['loss'][-1]
    train_acc = history.history['accuracy'][-1]

    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_test_classes = np.argmax(y_test, axis=1)

    optimizer_name = optimizer.__class__.__name__
    print(f"Batch Size: {batch_size}, Filter Size: {filter_size}, Optimizer: {optimizer_name}, "
          f"Regularization: {reg}, Train Accuracy: {train_acc:.4f}, Train Loss: {train_loss:.4f}, "
          f"Test Accuracy: {test_acc:.4f}, Test Loss: {test_loss:.4f}")

    plot_loss_accuracy(history)
    plot_confusion_matrix(y_test_classes, y_pred_classes)
    print(classification_report(y_test_classes, y_pred_classes))

    return batch_size, filter_size, optimizer_name, reg, train_acc, train_loss, test_acc, test_loss

# Define different hyperparameter combinations
hyperparams = [
    (200, 3, regularizers.l2(0.0001), optimizers.AdamW(learning_rate=0.0005), 150),
    (128, 3, regularizers.l2(0.0001), optimizers.AdamW(learning_rate=0.0005), 150)
]

# Store results
results = []
for params in hyperparams:
    results.append(train_and_evaluate(*params))

# Save results to CSV
df = pd.DataFrame(results, columns=['Batch Size', 'Filter Size', 'Optimizer', 'Regularization', 'Train Accuracy', 'Train Loss', 'Test Accuracy', 'Test Loss'])
df.to_csv('cnn_fashion_mnist_results.csv', index=False)
print("Results saved to cnn_fashion_mnist_results.csv")
```
